```python
import tensorflow as tf
import numpy as np
```

```python
filenames = ["../Datasets/Images/20.jpg", "../Datasets/Images/12.jpg", "../Datasets/Images/13.jpg", "../Datasets/Images/14.jpg", "../Datasets/Images/5.jpg", "../Datasets/Images/6.jpg", "../Datasets/Images/7.jpg", "../Datasets/Images/8.jpg", "../Datasets/Images/9.jpg", "../Datasets/Images/10.jpg"]
class_names = [1, 2, 3, 4, 5, 6, 7, 8, 9, 0]
```

```python
model = tf.keras.models.load_model("saved_model")
```

```python
index = 3
img_path = filenames[index]
img = tf.keras.preprocessing.image.load_img(img_path, target_size=(224, 224))
img_array = tf.keras.preprocessing.image.img_to_array(img)
img_array = tf.expand_dims(img_array, 0)
prediction = model.predict(img_array)
score = tf.nn.softmax(prediction[0])
label = class_names[np.argmax(prediction)]
print("This image most likely is a {} digit with {:.2f}% accuracy".format(class_names[np.argmax(score)], 100 * np.max(score)))
img
```
