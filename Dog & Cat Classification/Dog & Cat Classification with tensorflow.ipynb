```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
filenames = ["1.jpg", "2.jpg", "3.jpg", "4.jpg", "5.jpg", "6.jpg", "7.jpg", "8.jpg", "9.jpg", "10.jpg", "11.jpg", "12.jpg", "13.jpg", "14.jpg", "15.jpg", "16.jpg", "17.jpg", "18.jpg", "19.jpg"]
class_names = ["Cat", "Dog"]
```

```python
# model = tf.keras.models.load_model("Dog & Cat Class CNN 224")
model = tf.keras.models.load_model("saved_model")
```

```python
index = 18
img_path = filenames[index]
img = tf.keras.preprocessing.image.load_img(img_path, target_size=(86, 86))
img_array = tf.keras.preprocessing.image.img_to_array(img)
img_array = tf.expand_dims(img_array, 0)
img
```

```python
prediction = model.predict(img_array)
score = tf.nn.softmax(prediction[0])
label = class_names[np.argmax(prediction)]
"This image most likely is a {} with {:.2f}%".format(class_names[np.argmax(score)], 100 * np.max(score))
```
